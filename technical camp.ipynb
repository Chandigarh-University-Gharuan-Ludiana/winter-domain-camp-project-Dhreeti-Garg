```python
import pandas as pd
import matplotlib.pyplot as plt

# Upload your CSV file
from google.colab import files

uploaded = files.upload()  # Upload your CSV file

# Read the uploaded CSV file
csv_filename = list(uploaded.keys())[0]  # Automatically grabs the uploaded file name
df = pd.read_csv(csv_filename)

# Ensure the necessary columns exist
required_columns = ["Latitude", "Longitude", "Temperature"]
for column in required_columns:
    if column not in df.columns:
        raise ValueError(f"Missing required column: {column} in the CSV file.")

# Visualize the data as a scatter plot with a heatmap
plt.figure(figsize=(10, 6))
scatter = plt.scatter(
    df["Longitude"],
    df["Latitude"],
    c=df["Temperature"],
    cmap="hot",
    s=20,
    alpha=0.7
)
plt.title("Urban Heat Map (Temperature Data)")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.colorbar(scatter, label="Temperature (°C)")
plt.show()
```
